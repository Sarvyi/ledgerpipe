```python
transfer_keywords = [
    "transfer",
    "self",
    "upi to",
    "imps to",
    "neft to",
    "to own account",
    "internal",
]

df["is_transfer"] = df["Narration"].str.lower().str.contains('|'.join(transfer_keywords))
```

```python
df["is_internal"] = df["is_transfer"]
```

```python
cc_keywords = ["credit card", "cc payment", "card payment"]
df["is_cc_payment"] = df["Narration"].str.lower().str.contains("|".join(cc_keywords))
```

```python
wallet_keywords = ["paytm add", "gpay add", "wallet add", "mutual fund", "sip", "zerodha"]
df["is_wallet_or_invest"] = df["Narration"].str.lower().str.contains("|".join(wallet_keywords))
```

```python
import datetime
import pandas as pd

today = datetime.date.today()
first_day_this_month = today.replace(day=1)
last_month_end = first_day_this_month - datetime.timedelta(days=1)
last_month_start = last_month_end.replace(day=1)

start = pd.Timestamp(last_month_start).date()
end = pd.Timestamp(last_month_end).date()

mask = (df["Date"] >= start) & (df["Date"] <= end)

last_month_df = df[mask]
print(last_month_df)
```

```python
df["actual_spend"] = df["Withdrawal Amt."].where(df["Withdrawal Amt."] > 0)
total_spent_last_month = last_month_df["actual_spend"].sum()
total_spent_last_month
```
